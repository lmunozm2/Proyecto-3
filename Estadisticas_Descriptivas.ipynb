```python
materias = ['punt_ingles', 'punt_matematicas', 'punt_c_naturales', 'punt_sociales_ciudadanas', 'punt_lectura_critica']

# Configurar la visualización
plt.figure(figsize=(10, 6))

# Iterar sobre las columnas de las materias y dibujar un histograma para cada una
for i, materia in enumerate(materias):
    plt.hist(df[materia], bins=10, alpha=0.5, label=materia.replace('punt_', '').replace('_', ' '), edgecolor='black')

# Agregar leyenda
plt.legend()

# Configurar título y etiquetas de los ejes
plt.title('Distribución de Puntajes por Secciones')
plt.xlabel('Puntaje')
plt.ylabel('Frecuencia')
plt.grid(True)

# Mostrar el histograma
plt.show()
```

```python
# Calcular el promedio de puntajes para cada materia y clasificación
promedios_clasificacion = df.groupby('clasificacion')[materias].mean().transpose()

# Configurar la visualización
plt.figure(figsize=(12, 6))

# Graficar las líneas para cada materia
for materia in promedios_clasificacion.columns:
    plt.plot(promedios_clasificacion.index, promedios_clasificacion[materia], marker='o', label=materia)

# Configurar título y etiquetas de los ejes
plt.title('Promedio de Puntajes por Clasificación')
plt.xlabel('Materia')
plt.ylabel('Promedio de Puntaje')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejorar la legibilidad
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))

# Iterar sobre las columnas de las materias y dibujar boxplots separados para cada clasificación
for i, materia in enumerate(materias, start=1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='clasificacion', y=materia, data=df, palette='Set2')
    plt.title(materia.replace('punt_', '').replace('_', ' '))

# Ajustar el diseño de los subplots
plt.tight_layout()

# Mostrar los boxplots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular el promedio de puntajes para cada materia y clasificación
promedios = df.groupby('clasificacion')[materias].mean().reset_index()

# Configurar la visualización
plt.figure(figsize=(15, 8))

# Iterar sobre las columnas de las materias y dibujar barras separadas para cada clasificación
for i, materia in enumerate(materias, start=1):
    plt.subplot(2, 3, i)
    sns.barplot(x='clasificacion', y=materia, data=promedios, palette='Set2')
    plt.title(materia.replace('punt_', '').replace('_', ' '))

# Ajustar el diseño de los subplots
plt.tight_layout()

# Mostrar las barras
plt.show()
```
